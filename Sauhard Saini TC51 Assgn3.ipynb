Calculating central tendency using built-in functions

```python
mean_income = df['Income'].mean()
print('Mean Income: ', mean_income)

median_income = df['Income'].median()
print('Median Income: ', median_income)

max_income = df['Income'].max()
print('Max Income: ', max_income)

min_income = df['Income'].min()
print('Min Income: ', min_income)

std_income = df['Income'].std()
print('SD of Income: ', std_income)

var_income = df['Income'].var()
print('Variance Income: ', var_income)
```

Calculating central tendency using without built-in functions

```python
arr = list(df['Income'])

def mean(arr):
    return sum(arr)/len(arr)

def median(arr):
    if len(arr)%2 == 0:
        arr = sorted(arr)
        return (arr[len(arr)//2] + arr[len(arr)//2 - 1]) / 2
    else:
        arr = sorted(arr)
        return arr[len(arr)//2]

def maximum(arr):
    return max(arr)

def minimum(arr):
    return min(arr)

def standard_deviation(arr):
    m = mean(arr)
    N = len(arr)
    return (sum([(i-m)**2 for i in arr])/(N-1))**0.5

def variance(arr):
    m = mean(arr)
    N = len(arr)
    return (sum([(i-m)**2 for i in arr])/(N-1))

mean_income = mean(arr)
print('Mean Income: ', mean_income)

median_income = median(arr)
print('Median Income: ', median_income)

max_income = maximum(arr)
print('Max Income: ', max_income)

min_income = minimum(arr)
print('Min Income: ', min_income)

std_income = standard_deviation(arr)
print('Standard Deviation Income: ', std_income)

var_income = variance(arr)
print('Variance Income: ', var_income)
```

```python
df = pd.read_csv('Iris.csv')
df
```

Calculating the basic stats and dropping the id column

```python
species = df['Species'].unique()
df.drop(columns=['Id'], inplace=True)
for s in species:
    print('Species: ', s)
    print(df[df['Species']==s].describe())
    print()
```

Correlation using scatterplot

```python
figure = plt.figure(figsize=(15, 15))
index = 1
for i in range(3):
    for j in range(i+1,4):
        plt.subplot(3, 2, index)
        plt.scatter(df[df.columns[i]], df[df.columns[j]], color='green')
        plt.xlabel(df.columns[i])
        plt.ylabel(df.columns[j])
        index += 1
```
