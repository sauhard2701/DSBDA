Importing libraries and dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# set a grey background 
sns.set(style="darkgrid")
df = sns.load_dataset("iris")

# list all the features and their types
print(df.dtypes)
```

```python
figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="sepal_length", hue="species")
plt.show()

figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="sepal_length")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="sepal_width",hue="species")
plt.show()

figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="sepal_width")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="petal_length",hue="species")
plt.show()

figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="petal_length")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="petal_width",hue="species")
plt.show()

figure = plt.figure(figsize=(11, 5))
sns.histplot(data=df, x="petal_width")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.boxplot(data=df, x='species', y="sepal_length",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.boxplot(data=df, x='species', y="sepal_width",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.boxplot(data=df, x='species', y="petal_length",hue="species")
plt.show()
```

```python
figure = plt.figure(figsize=(11, 5))
sns.boxplot(data=df, x='species', y="petal_width",hue="species")
plt.show()
```

Compare distributions and identify outliers.
1. In fig.1 we can see that sepal length of setosa< Versicolor < Virginica and one value is outlier in Virginica.
